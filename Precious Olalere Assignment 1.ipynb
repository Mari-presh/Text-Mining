```python
#Using NLTK to tokenize the statement text (statement & cleaned_statement column). 
#nltk.download('punkt')
from nltk.tokenize import word_tokenize
train["tokenized_statement"] = train["statement"].apply(word_tokenize)
train["tokenized_cleaned_statement"] = train["cleaned_statement"].apply(word_tokenize)
train[["statement","tokenized_statement","cleaned_statement","tokenized_cleaned_statement"]].head(10)
```

```python
#Using another package (e.g., spacy) or your own code to tokenize the statement text.

nlp = spacy.load('en_core_web_sm')
def tokenize_spacy(text):
    # Process the text with spaCy
    doc = nlp(text)
    # Extract tokens
    return [token.text for token in doc]

train['tokenized_statement_spacy'] = train['statement'].apply(tokenize_spacy)
```

```python
train['tokenized_cleaned_statement_spacy'] = train['cleaned_statement'].apply(tokenize_spacy)
```

```python
#Using NLTK to remove stopwords from the tokenized_cleaned_statement and create column named stopword_removed_statement.
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
def remove_stopwords(tokenized_text):
    return [word for word in tokenized_text if word not in stop_words]

train["stopword_removed_statement"] = train["tokenized_cleaned_statement"].apply(remove_stopwords)
train[["tokenized_cleaned_statement", "stopword_removed_statement"]].head(10)
```

```python
#Using Porter Stemmer (implemented in NLTK) to stem the text in stopword_removed_statement and 
#create a new column called stemmed_statement.

from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()

def stem_words(tokenized_text):
    return [stemmer.stem(word) for word in tokenized_text]
train["stemmed_statement"] = train["stopword_removed_statement"].apply(stem_words)
train[["stopword_removed_statement", "stemmed_statement"]].head(10)
```

```python
#Using another stemmer included in NLTK to stem the stopword_removed_statement text. 
#Write about their similarities and differences.
#Using the snowball stemmer

from nltk.stem import SnowballStemmer
```

```python
# Importing the Snowball Stemmer for English
snowball_stemmer = SnowballStemmer("english")

def snowball_stem_words(tokenized_text):
    return [snowball_stemmer.stem(word) for word in tokenized_text]

train["snowball_stemmed_statement"] = train["stopword_removed_statement"].apply(snowball_stem_words)

train[["stopword_removed_statement", "snowball_stemmed_statement"]].head(10)
```
