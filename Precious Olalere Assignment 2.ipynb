```python
#Task 1: Load the dataset
#List down the number of statements in the training set and the test set

train_liar = pd.read_csv("LIAR_train.csv")
test_liar = pd.read_csv("LIAR_test.csv")


print ("Train set: ",len(train_liar))
print ("Test set: ",len(test_liar))

train_liar.head(10)
```

```python
# enriching the statement for train dataset
def enrich_statement(row):
    return (f"{row['statement']} The topic of the statement is {row['topic']}. "
            f"The speaker is {row['speaker']}. The location is {row['location']}. "
            f"The party is {row['party']}. The statement comes from a {row['context']}.")

train_liar['enriched_statement'] = train_liar.apply(enrich_statement, axis=1)
pd.set_option('display.max_colwidth', None)
train_liar[["statement", "enriched_statement"]].head(5)
```

```python
# Enriching the statement for test dataset
def enrich_statement(row):
    return (f"{row['statement']} The topic of the statement is {row['topic']}. "
            f"The speaker is {row['speaker']}. The location is {row['location']}. "
            f"The party is {row['party']}. The statement comes from a {row['context']}.")

test_liar['enriched_statement'] = test_liar.apply(enrich_statement, axis=1)
test_liar[["statement", "enriched_statement"]].head(5)
```

```python
#Task 3: Extract unigram features
from sklearn.feature_extraction.text import CountVectorizer

train_text = train_liar["enriched_statement"]
test_text = test_liar["enriched_statement"]

# set the n-gram range
vectorizer = CountVectorizer(ngram_range = (1,1))

# create training data representation
train_data_cv = vectorizer.fit_transform(train_text.values.astype('U'))
print(train_data_cv.shape,"\n") 

# create test data representation
test_data_cv = vectorizer.transform(test_text.values.astype('U'))
print(test_data_cv.shape,"\n") 



```

```python
#Task 4: Train and evaluate classifier

from sklearn.naive_bayes import MultinomialNB
# define true labels from train set
x_train = train_data_cv
y_train = train_liar["label"]
x_test = test_data_cv
y_test = test_liar["label"]

# build model on the training data
model = MultinomialNB()
model.fit(x_train, y_train)

# predict the labels for the test data
predictions = model.predict(x_test)
```

```python
import pickle

# save the model to disk
filename = 'NaiveBayesModelBOW.sav'
pickle.dump(model, open(filename, 'wb'))

# some time later...
# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(x_test, y_test)
print(result)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report, confusion_matrix 

print ("Accuracy score: ", accuracy_score(y_test, predictions))
print ("Individual label performance: ")
print (classification_report(y_test, predictions))
print (confusion_matrix(y_test, predictions))
```

```python
#Task 8:  Preprocess the data and apply the best-performing model
#Preprocess the statement column using one of the text preprocessing methods introduced earlier
#(e.g. stemming, stopword removal). You can store the preprocessed text in a new column. 
#Using NLTK to tokenize the enriched statement for train dataset(enriched statement & tokenized_enriched_statement column). 

#to carry out stop wprd removal tokenization has to been done first. 
import nltk
from nltk.tokenize import word_tokenize
train_liar["tokenized_enriched_statement"] = train_liar["enriched_statement"].apply(word_tokenize)
train_liar[["enriched_statement","tokenized_enriched_statement"]].head(10)
```

```python
#Using NLTK to tokenize the enriched statement for test dataset(enriched statement & tokenized_enriched_statement column). 

import nltk
from nltk.tokenize import word_tokenize
test_liar["tokenized_enriched_statement"] = test_liar["enriched_statement"].apply(word_tokenize)
test_liar[["enriched_statement","tokenized_enriched_statement"]].head(10)
```

```python
#removing stopword from train dataset 
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
def remove_stopwords(tokenized_text):
    return [word for word in tokenized_text if word not in stop_words]

train_liar["stopword_removed_statement"] = train_liar["tokenized_enriched_statement"].apply(remove_stopwords)
train_liar[["tokenized_enriched_statement", "stopword_removed_statement"]].head(10)
```

```python
#removing stopword from test dataset 
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
def remove_stopwords(tokenized_text):
    return [word for word in tokenized_text if word not in stop_words]

test_liar["stopword_removed_statement"] = test_liar["tokenized_enriched_statement"].apply(remove_stopwords)
test_liar[["tokenized_enriched_statement", "stopword_removed_statement"]].head(10)
```

```python
#Based on the model performance from task 3, 4, 5,. 
#The unigram model is the best performing model, therefore, I will be using it for the preprocessed text. 

#Extract unigram features
from sklearn.feature_extraction.text import CountVectorizer

train_text = train_liar["stopword_removed_statement"].apply(lambda x: " ".join(x))
test_text = test_liar["stopword_removed_statement"].apply(lambda x: " ".join(x))

# set the n-gram range
vectorizer = CountVectorizer(ngram_range = (1,1))

# create training data representation
train_data_cv = vectorizer.fit_transform(train_text.values.astype('U'))
print(train_data_cv.shape,"\n") 

# create test data representation
test_data_cv = vectorizer.transform(test_text.values.astype('U'))
print(test_data_cv.shape,"\n") 

```

```python
#Train and evaluate classifier

from sklearn.naive_bayes import MultinomialNB
# define true labels from train set
x_train4 = train_data_cv
y_train4 = train_liar["label"]
x_test4 = test_data_cv
y_test4 = test_liar["label"]

# build model on the training data
model = MultinomialNB()
model.fit(x_train4, y_train4)

# predict the labels for the test data
predictions4 = model.predict(x_test4)
```

```python
#save the model to disk
filename = 'NaiveBayesModelBOW.sav'
pickle.dump(model, open(filename, 'wb'))

# some time later...
# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(x_test4, y_test4)
print(result)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report, confusion_matrix 

print ("Accuracy score: ", accuracy_score(y_test4, predictions4))
print ("Individual label performance: ")
print (classification_report(y_test4, predictions4))
print (confusion_matrix(y_test4, predictions4))
```

```python
#Task 11: Three-way vs six-way classification

# Assuming your original labels are in a column named 'label'
# Re-map labels to three categories: Negative, Neutral, Positive

group_label = {
    "pants-fire": "negative",
    "false": "negative",
    "barely-true": "neutral",
    "half-true": "neutral",
    "mostly-true": "positive",
    "true": 'positive'
}

# Apply the mapping to train and test dataset
train_liar["three_way"] = train_liar["label"].map(group_label)
test_liar["three_way"] = test_liar["label"].map(group_label)

```

```python
from sklearn.feature_extraction.text import CountVectorizer

train_textnew = train_liar["stopword_removed_statement"].apply(lambda x: " ".join(x))
test_textnew = test_liar["stopword_removed_statement"].apply(lambda x: " ".join(x))

# set the n-gram range
vectorizer = CountVectorizer(ngram_range = (1,1))

# create training data representation
train_data_cv = vectorizer.fit_transform(train_textnew.values.astype('U'))
print(train_data_cv.shape,"\n") 

# create test data representation
test_data_cv = vectorizer.transform(test_textnew.values.astype('U'))
print(test_data_cv.shape,"\n") 
```

```python
#Train and evaluate classifier

from sklearn.naive_bayes import MultinomialNB
# define true labels from train set
x_train5 = train_data_cv
y_train5 = train_liar["three_way"]
x_test5= test_data_cv
y_test5 = test_liar["three_way"]

# build model on the training data
model = MultinomialNB()
model.fit(x_train5, y_train5)

# predict the labels for the test data
predictions5 = model.predict(x_test5)
```

```python
#save the model to disk
filename = 'NaiveBayesModelBOW.sav'
pickle.dump(model, open(filename, 'wb'))

# some time later...
# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(x_test5, y_test5)
print(result)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report, confusion_matrix 

print ("Accuracy score: ", accuracy_score(y_test5, predictions5))
print ("Individual label performance: ")
print (classification_report(y_test5, predictions5))
print (confusion_matrix(y_test5, predictions5))
```

The three way classification perfomed better than the six way classification. The f1 score of the six-way classification was so low compared to the three-way classification as well as the prediction accuracy.
